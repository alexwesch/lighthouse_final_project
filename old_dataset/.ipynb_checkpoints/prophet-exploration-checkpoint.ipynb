```python
import itertools
from prophet.diagnostics import performance_metrics
```

```python
param_grid = {
    'changepoint_prior_scale': [0.0001, 0.001, 0.01],
    'seasonality_prior_scale': [.001, 0.01, 0.1],
    'holidays_prior_scale': [0.01, 0.1, 1],
    'seasonality_mode': ['additive', 'multiplicative']
}
```

```python
# Generate all combinations of parameters
all_params = [dict(zip(param_grid.keys(), v)) for v in itertools.product(*param_grid.values())]
rmses = []  # Store the RMSEs for each params here
```

```python
# Use cross validation to evaluate all parameters
for params in all_params:
    m = Prophet(**params).fit(sales)  # Fit model with given params
    df_cv = cross_validation(m, horizon='30 days', parallel="processes")
    df_p = performance_metrics(df_cv, rolling_window=1)
    rmses.append(df_p['rmse'].values[0])
```

```python
best_params = all_params[np.argmin(rmses)]
print(best_params)
```

```python
model = Prophet(interval_width=0.95, changepoint_prior_scale=.001, seasonality_prior_scale=.01)
model.add_seasonality('yearly', period=365, fourier_order=10, mode='additive') #default 10
model.fit(train)
y_forecast = model.make_future_dataframe(periods=1, freq='W')
y_forecast = model.predict(y_forecast)
plt.figure(figsize=(18, 6))
model.plot(y_forecast, xlabel = 'Date', ylabel = 'Mock Regular')
plt.title('Mock Regular Predictions');
```
